```python
import sqlite3
import pandas as pd
```

```python
with sqlite3.connect('county_36hr_forecast.sqlite') as conn:
    df = pd.read_sql_query('SELECT * FROM county', conn)

df.head(10)
```

```python
rain = df.loc[df['element'] == 'PoP', ['location', 'issueTime', 'startTime', 'endTime', 'value']]

# dealing with time zone
for time_type in ['issueTime', 'startTime', 'endTime']:
    rain[time_type] = pd.to_datetime(rain[time_type]).dt.tz_localize(None)
    
rain['lead_time'] = rain['startTime'] - rain['issueTime']
rain.head()
```
